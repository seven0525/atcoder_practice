```python
if n[0] + n[1] + n[2] + n[3] == 7:
    print("{}+{}+{}+{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] + n[1] + n[2] - n[3] == 7:
    print("{}+{}+{}-{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] + n[1] - n[2] + n[3] == 7:
    print("{}+{}-{}+{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] + n[1] - n[2] - n[3] == 7:
    print("{}+{}-{}-{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] - n[1] + n[2] + n[3] == 7:
    print("{}-{}+{}+{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] - n[1] + n[2] - n[3] == 7:
    print("{}-{}+{}-{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] - n[1] - n[2] + n[3] == 7:
    print("{}-{}-{}+{}=7".format(n[0],n[1],n[2],n[3]))
elif n[0] - n[1] - n[2] - n[3] == 7:
    print("{}-{}-{}-{}=7".format(n[0],n[1],n[2],n[3]))
else:
    print("error")
```

```python
a,b,c,d = list(map(int, list(str(input()))))

def calc(n, op, m):
    if(type(n) is not int):
        return False
    if(op == "+"):
        return n + m
    elif(op == "-"):
        return n - m
    elif(op == "*"):
        return n * m
    elif(m != 0):
        return n / m
    else:
        return False

ops1 = ["/", "+", "-", "*"]
ops2 = ["/", "+", "-", "*" ]
ops3 = ["/", "+", "-", "*" ]
for op1 in ops1:
    for op2 in ops2:
        for op3 in ops3:
            if(calc(calc(calc(a,op1,b),op2,c),op3,d) == 7):
        print(a,op1,b,op2,c,op3,d,"=7", sep="")
        exit()
```

# C - All Green
https://atcoder.jp/contests/abc104/tasks/abc104_c

```python
d,g = map(int, input().split())
points = []
bonus = []
for i in range(d):
    p,b = map(int, input().split())
    points.append(p)
    bonus.append(b)
```

```python
import math

ans = []
for n in range(len(points)):
    need_num = (g / ((n+1) * 100))
    need_num = math.ceil(need_num)
    if need_num < points[n]:
        ans.append(need_num)
    elif (n+1) * 100 * points[n] + bonus[n] >= g:
        ans.append(points[n])
print(min(ans))
```
